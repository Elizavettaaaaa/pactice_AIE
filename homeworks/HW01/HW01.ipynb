```python
# Импортируем библиотеку NumPy
import numpy as np
```

```python
# 1. Создание массива из обычного списка
my_list = [10, 20, 30, 40, 50]
arr_list = np.array(my_list)

# 2. Создание массива из нулей (3 строки, 4 столбца)
arr_zeros = np.zeros((3, 4))

# 3. Создание массива чисел от 0 до 9 (функция arange)
arr_range = np.arange(10)

print("Массив из списка:", arr_list)
print("Массив нулей:\n", arr_zeros)
print("Массив arange:", arr_range)
```

```python
# Показываем форму и тип данных для массива arr_zeros
print(f"Форма массива (shape): {arr_zeros.shape}") # (3, 4) -> 3 строки, 4 столбца
print(f"Тип данных (dtype): {arr_zeros.dtype}")    # float64
```

```python
# Получение 3-го элемента из arr_list (индексы начинаются с 0)
element = arr_list[2]
print("Третий элемент (индекс 2):", element)

# Срез (извлечение элементов с 1-го по 4-й не включительно)
sub_array = arr_list[1:4]
print("Срез [1:4]:", sub_array)
```

```python
# Поэлементное умножение массива на 2
arr_mult = arr_list * 2
print("Массив, умноженный на 2:", arr_mult)

# Сложение двух массивов
arr_sum = arr_list + arr_mult
print("Сумма массивов:", arr_sum)

# Вычисление среднего значения (mean)
arr_mean = np.mean(arr_list)
print("Среднее значение:", arr_mean)

# Поиск максимального значения
arr_max = np.max(arr_list)
print("Максимальное значение:", arr_max)
```
